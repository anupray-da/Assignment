11. What is memory management in Python?
 - Python utilizes automatic memory management, primarily through a mechanism called garbage collection. This means you, as the programmer, generally don't need to manually allocate or deallocate memory for objects. Python handles it behind the scenes.
 - Python objects and data structures are stored in a private heap space. This heap is managed internally by the Python memory manager.
 - Python keeps track of how many references (or pointers) exist to each object. When an object's reference count drops to zero, it means the object is no longer accessible and can be garbage collected.
 - The garbage collector periodically scans the heap for objects with a reference count of zero. These objects are then deallocated, freeing up memory for reuse.
 - Python also has a mechanism to detect and collect objects involved in cyclic references.

15. How does Python's garbage collection system work?
 - Python's garbage collection (GC) system is responsible for automatically managing memory, reclaiming it when objects are no longer in use. This helps developers avoid manual memory management and reduces memory leaks. Python uses a combination of reference counting and a cyclic garbage collector to manage memory.
 - reference counting- Python primarily uses reference counting to keep track of how many references exist to an object. Each object in Python has an associated reference count. When the reference count drops to zero, the object is deallocated.
 - Cyclic Garbage Collection- One limitation of reference counting is that it cannot handle reference cycles—situations where objects reference each other, preventing their reference counts from reaching zero.

```python
#18. How would you implement a basic logging setup that logs to a file with rotation after 1MB?
def logging_setup(log_file):
    import logging
    from logging.handlers import RotatingFileHandler

    logger = logging.getLogger(__name__)
    logger.setLevel(logging.INFO)
    handler = RotatingFileHandler(log_file, maxBytes=1000000, backupCount=5)
    formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
    handler.setFormatter(formatter)
    logger.addHandler(handler)
    return logger
```

```python
logging_setup("log.txt")
```

```python
#21. Write a Python program that reads a file and prints the number of occurrences of a specific word.
def count_occ(file_name, word):
    try:
        with open(file_name, 'r') as file:
            content = file.read()
            content = content.lower()
            word = word.lower()
            occurrences = content.count(word)
            return occurrences
            print(f"The word '{word}' appears {occurrences} times in the file.")
    except FileNotFoundError:
        print(f"Error: File '{file_name}' not found.")
```

```python
count_occ("test.txt", "hello")
```
