5. What are iterators in Python and how do they differ from iterables?

6. Explain the concept of generators in Python and how they are defined.
 - Generators are type of function which produces the values one at a time instead of all at once, which saves memory.
 - It uses a keyword called 'yield' which pauses the execution and provide a single value to caller.
 - It generally used in large datasets and provides results when needed which improves the efficiency.

```python
#Ex of generator
a=[1,2,3]
def generat(a):
 for i in range(a):
  yield i**2
gen=generat(5) # here 'gen' is generator
print(next(gen))
```

7. What are the advantages of using generators over regular functions?
 - Generators are normally produces the values one at a time instead of all at once so we dont need all values to store at once. Where as regular function returns all the value at once which consumes more memory and take more time.
 - Generators generally used in large datasets and it provides results when needed which make it more efficient. But regular function is less efficient than generators as it provides all the results.
 - Generators can be used for infinite sequences but regular functions cant't be used.

```python
#Ex of Generators over regular function
def generat(a):
  for i in range(a):
    yield i**2
gen=generat(10000000) # here we can generate one by by using next
print(next(gen))

def regular(a):
  result=[]
  for i in range(a):
    result.append(i**2)
  return result
regular(10000000) #here it will take more memory and time as all results going to print at once
```

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
 - The `map()` function executes a specified function for each item in an iterable. The item is sent to the function as a parameter.
 - syntax:- map(function,iterable)
 -`reduce()` - Applies a function of two arguments cumulatively to the items of an iterable, from left to right, to reduce the iterable to a single value.
 - We need to import it from 'functools' module.
 - syntax- functools.reduce(function, iterable[, initializer])
 - `filter()`- Constructs an iterator from those elements of an iterable for which a given function returns True.
 - It filters the elements of the iterable based on a condition specified by the function.
 - syntax- filter(function,iterable)

```python
x=[1,2,3,4,5,6]

#Ex of map() function
list(map(lambda a:a**2,x)) #square the numbers in the list

#Ex of reduce() function
from functools import reduce
reduce(lambda a,b:a+b,x) #sum of all the numbers in the list

#Ex of filter() function
list(filter(lambda a:a%2==0,x)) #filter the even numbers from the given list
```

```python
# 4)Write a Python function that checks if a given number is prime or not from 1 to 200.
def prime(n):
  if n<2:
    print('It is not a prime Number')
    return
  for i in range(2,int(n**0.5)+1):
    if n%i==0:
      print('It is not a prime Number')
      return
  print('It is a prime Number')
```

```python
# 5)Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class Fibon:
  def __init__(self, n):
        self.n = n
        self.a, self.b = 0, 1
        self.count = 0
  def __iter__(self):
        return self
  def __next__(self):
        if self.count < self.n:
            result = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return result
        else:
            raise StopIteration
for i in Fibon(10):
  print(i)
```

```python
# 7)Implement a generator function that reads a file line by line and yields each line as a string.
def read_file(filename):
  with open(filename,'r') as file:
    for line in file:
      yield line

with open('file.txt','w') as file:
  file.write('My name is Anup\n')
  file.write('My age is 26\n')
  file.write('This is Function Assignment\n')

for line in read_file('file.txt'):
  print(line)
```
